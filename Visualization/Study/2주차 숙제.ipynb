```python
from konlpy.tag import Okt
okt = Okt()

import re
from collections import Counter
```

```python
path = '/content/drive/MyDrive/BOAZ 멘멘조 스터디/Music_melon.csv'
df = pd.read_csv(path)
df.head()
```

```python
df.isnull().sum()
```

```python
# 1. 클리닝 (Text Cleaning)
goal_list = []
for k in range(len(df['text'])):
    goal = re.sub(r'[^\s\w]', ' ', df['text'][k]).replace('\n', ' ') # cleaning
    goal_list.append(goal)
```

```python
df['preprocessed_text'] = goal_list
df
```

```python
df_IU = df[df['artist'] == '아이유']
df_IU
```

```python
goal_text = ' '.join(df_IU['preprocessed_text'])
print(goal_text)

# 클리닝 + 스테밍 + 품사 필터링

# 1. 클리닝 (불필요한 문장 부호, 이모티콘, 특수문자 제거)
goal_text = re.sub(r'[^\s\w]', ' ', goal_text) # 1. 클리닝


# 2. 토크나이징 (주어진 문장에서 형태소 단위로 정보를 나누기) 3. 스테밍 (단어의 원형 또는 어간 찾기)
def pos_filtering(text): 
    pos_word_list = okt.pos(text, stem = True) #okt.pos 주어진 텍스트를 형태소 단위로 나누고, 나눠진 각 형태소를 그에 해당하는 품사와 함께 리스트화
    pos_list = ['Noun', 'Verb', 'Adjective']
    pos_filtered_word_list = []

 # 4. 품사 태킹 (단어가 문장에서 어떤 품사인지 표시하기)
    for word, pos in pos_word_list:
        if pos in pos_list:
            pos_filtered_word_list.append(word)

    return pos_filtered_word_list
```

```python
# 단어 리스트 생성
word_list = pos_filtering(goal_text)
print('총 {}개의 단어'.format(len(word_list)))
```

```python
# 명사만
word_list = okt.nouns(goal_text)
len(word_list)
```

```python
new_word_list = []
for word in word_list:
  if len(word) != 1:
    new_word_list.append(word)

len(new_word_list)
```

```python
c = Counter(new_word_list)

num = 150 # 상위 150개 단어만 출력
print(c.most_common(num))
```

```python
word_df = pd.DataFrame(c.most_common(num), columns = ['word', 'count']); 
word_df
```
